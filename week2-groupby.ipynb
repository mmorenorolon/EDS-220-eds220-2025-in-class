```python
import numpy as np
import pandas as pd

# Load Palmer penguins data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

# Check In
a. Use the max() method for pandas.DataFrames to calculate the maximum value of a penguinâ€™s body mass by year and species.

```python
penguins.groupby(['year', 'species']).body_mass_g.max()
```

b. Use (1) to display the highest body masses per year and species as a bar plot in descending order.

```python
penguins.groupby(['year', 'species']).body_mass_g.max().sort_values().plot(kind = 'barh')
```
